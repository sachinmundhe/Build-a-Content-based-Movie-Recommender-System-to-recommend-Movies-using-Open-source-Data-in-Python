```python
df_new=df_final.copy()
```

We will now create a new column called 'rating_between'. We will divide our average rating column into various categories such as rating between 0 and 1, 1 and 2 and so on. This will work as one of the features that we will feed to our model so that it can make better predictions.

```python
df_new.loc[ (df_new['average_rating'] >= 0) & (df_new['average_rating'] <= 1), 'rating_between'] = "between 0 and 1"
df_new.loc[ (df_new['average_rating'] > 1) & (df_new['average_rating'] <= 2), 'rating_between'] = "between 1 and 2"
df_new.loc[ (df_new['average_rating'] > 2) & (df_new['average_rating'] <= 3), 'rating_between'] = "between 2 and 3"
df_new.loc[ (df_new['average_rating'] > 3) & (df_new['average_rating'] <= 4), 'rating_between'] = "between 3 and 4"
df_new.loc[ (df_new['average_rating'] > 4) & (df_new['average_rating'] <= 5), 'rating_between'] = "between 4 and 5"
```

```python
rating_df = pd.get_dummies(df_new['rating_between'])
rating_df.head()
```
